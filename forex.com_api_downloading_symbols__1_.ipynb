```python
import time

def extract_market_info(api_result, market_data):
    for market in api_result:
        market_id = market['MarketId']
        name = market['Name']
        if {'MarketId': market_id, 'Name': name} not in market_data:
            market_data.append({'MarketId': market_id, 'Name': name})
            print(name)
        # If the market has its own markets, recursively call this function
        if 'Markets' in market:
            extract_market_info(market['Markets'], market_data)

# Initialize an empty list to store the results
market_data = []

# Loop over the symbol_list
for symbol in symbol_list:
    try:
        # Call the API
        api_result = api.get_market_info(symbol, get=None)['Markets']

        # Extract the market info
        extract_market_info(api_result, market_data)

    except IndexError:
        print(f"Error: Failed to access data for symbol '{symbol}'. List index out of range.")
    except Exception as e:
        print(f"Error: {str(e)} encountered for symbol '{symbol}'.")

    # Wait for 5 seconds before the next API call
    time.sleep(5)

print(market_data)
```

```python
df = pd.DataFrame(market_data)

# Reorder the DataFrame columns
df = df[['Name', 'MarketId']]

# Save the DataFrame as a CSV file
df.to_csv('forexdotcom.csv', index=False)
```
